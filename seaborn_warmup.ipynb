````python
sns.barplot( 'origin', 'horsepower', data = df, ci = None)
````

````python
sns.scatterplot('horsepower', 'displacement', data = df, hue = 'weight',palette = 'Greens')
````

````python
sns.scatterplot('mpg', 'displacement', data=df, hue = 'cylinders')
````

````python
df.cylinders = df.cylinders.astype('category')
````

````python
sns.scatterplot('mpg', 'displacement', data = df, hue = 'cylinders')
````

````python
sns.relplot(data = df, y = 'mpg', x = 'displacement',col = 'origin', hue = 'cylinders')
````

````python
fig, (ax1, ax2) = plt.subplots(1,2, figsize=(13,7))

sns.scatterplot(data=df, y='mpg', x = 'displacement', ax = ax1)
sns.scatterplot(data=df, y='mpg', x = 'weight', ax = ax2)


ax1.set(title = 'displacement vs mpg')
ax2.set(title = 'weight vs mpg')
fig.suptitle('cars mpg visualizations')
````
